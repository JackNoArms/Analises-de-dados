## Dicionários para referência no BD

```python
# Dicionários
ufs = {'Rondônia' : 11,
  'Acre' : 12,
  'Amazonas' : 13,
  'Roraima' : 14,
  'Pará' : 15,
  'Amapá' : 16,
  'Tocantins' : 17,
  'Tocantis' : 17,
  'Maranhão' : 21,
  'Piauí' : 22,
  'Ceará' : 23,
  'Rio Grande do Norte' : 24,
  'Paraíba' : 25,
  'Pernambuco' : 26,
  'Alagoas' : 27,
  'Sergipe' : 28,
  'Bahia' : 29,
  'Minas Gerais' : 31,
  'Espírito Santo' : 32,
  'Rio de Janeiro' : 33,
  'São Paulo' : 35,
  'Paraná' : 41,
  'Santa Catarina' : 42,
  'Rio Grande do Sul' : 43,
  'Mato Grosso do Sul' : 50,
  'Mato Grosso' : 51,
  'Goiás' : 52,
  'Distrito Federal' : 53}

ufs_s = {'RO' : 11,
  'AC' : 12,
  'AM' : 13,
  'RR' : 14,
  'PA' : 15,
  'AP' : 16,
  'TO' : 17,
  'MA' : 21,
  'PI' : 22,
  'CE' : 23,
  'RN' : 24,
  'PB' : 25,
  'PE' : 26,
  'AL' : 27,
  'SE' : 28,
  'BA' : 29,
  'MG' : 31,
  'ES' : 32,
  'RJ' : 33,
  'SP' : 35,
  'PR' : 41,
  'SC' : 42,
  'RS' : 43,
  'MS' : 50,
  'MT' : 51,
  'GO' : 52,
  'DF' : 53}

loc = {'Rural' : 2, 
       'Total' : 3, 
       'Urbana' : 1}

dep = {'Estadual' : 2,
      'Federal' : 1,
      'Municipal': 4,
      'Privada' : 8,
      'Pública' : 7,
      'Total' : 15}
```

## Tratamentos de dados do Dataframe df1

```python
# Substitui os valores da coluna 'UNIDGEO' pela coluna 'CD_IBGE_UF', dropa a coluna e substitui os valores pelo o do respectivo dicionário
df1['CD_IBGE_UF'] = df1['UNIDGEO']
df1.drop('UNIDGEO', axis=1, inplace=True)
df1['CD_IBGE_UF'].replace(ufs, inplace=True)

# Substitui os valores da coluna 'NO_DEPENDENCIA' pela coluna 'TP_DEPENDENCIA_ADM', dropa a coluna e substitui os valores pelo o do respectivo dicionário
df1['TP_DEPENDENCIA_ADM'] = df1['NO_DEPENDENCIA']
df1['TP_DEPENDENCIA_ADM'].replace(dep, inplace=True)
df1.drop('NO_DEPENDENCIA', axis=1, inplace=True)

# Substitui os valores da coluna 'NO_CATEGORIA' pela coluna 'CD_LOCALIZACAO_ESCOLA', dropa a coluna e substitui os valores pelo o do respectivo dicionário
df1['CD_LOCALIZACAO_ESCOLA'] = df1['NO_CATEGORIA']
df1['CD_LOCALIZACAO_ESCOLA'].replace(loc, inplace=True)
df1.drop('NO_CATEGORIA', axis=1, inplace=True)

# Renomeando a coluna 'NU_ANO_CENSO' para 'ANO_CENSO'
df1 = df1.rename(columns={'NU_ANO_CENSO': 'ANO_CENSO'})
df1
```

```python
# Listas de prefixos e sufixos
antigos_prefixos = ['FUN_CAT', 'FUN_AI_CAT', 'FUN_AF_CAT', 'MED_CAT']
novos_prefixos = ['EF', 'EF_AI', 'EF_AF', 'EM']
sufixos = ['N1', 'N2', 'N3', 'N4', 'N5', 'N6']
antigos_sufixos = ['1', '2', '3', '4', '5', '6']

# Loop para adicionar as novas colunas com os sufixos
for prefixo in antigos_prefixos:
    for antigo, novo in zip(antigos_sufixos, sufixos):
        # Adiciona a nova coluna com o sufixo "N" usando o novo prefixo
        df1[f'{novos_prefixos[antigos_prefixos.index(prefixo)]}_{novo}'] = df1[f'{prefixo}_{antigo}']
        
        # Remove a coluna antiga
        df1.drop(f'{prefixo}_{antigo}', axis=1, inplace=True)

# Lista de colunas a serem removidas (N3 a N6 e FUN_CAT_3 a FUN_CAT_6, FUN_AI_N3 a FUN_AI_N6, FUN_AF_N3 a FUN_AF_N6)
colunas_a_remover = [f'{novos_prefixos[antigos_prefixos.index(prefixo)]}_{novo}' for prefixo in antigos_prefixos for novo in sufixos[2:]]

# Remove as colunas não desejadas
df1.drop(colunas_a_remover, axis=1, inplace=True)

# Remove todas as linhas que não sejam numéricas
df1['CD_IBGE_UF'] = pd.to_numeric(df1['CD_IBGE_UF'], errors='coerce')
df1.dropna(subset=['CD_IBGE_UF'], inplace=True, axis=0)

# Convertendo para inteiro
df1['CD_IBGE_UF'] = df1['CD_IBGE_UF'].astype(int)
df1
```

## Tratamentos de dados do Dataframe df2

```python
# Substitui os valores da coluna 'SG_UF' pela coluna 'CD_IBGE_UF', dropa a coluna e substitui os valores pelo o do respectivo dicionário
df2['CD_IBGE_UF'] = df2['SG_UF']
df2.drop('SG_UF', axis=1, inplace=True)
df2['CD_IBGE_UF'].replace(ufs_s, inplace=True)

# Substitui os valores da coluna 'NO_DEPENDENCIA' pela coluna 'TP_DEPENDENCIA_ADM', dropa a coluna e substitui os valores pelo o do respectivo dicionário
df2['TP_DEPENDENCIA_ADM'] = df2['NO_DEPENDENCIA']
df2['TP_DEPENDENCIA_ADM'].replace(dep, inplace=True)
df2.drop('NO_DEPENDENCIA', axis=1, inplace=True)

# Substitui os valores da coluna 'NO_CATEGORIA' pela coluna 'CD_LOCALIZACAO_ESCOLA', dropa a coluna e substitui os valores pelo o do respectivo dicionário
df2['CD_LOCALIZACAO_ESCOLA'] = df2['NO_CATEGORIA']
df2['CD_LOCALIZACAO_ESCOLA'].replace(loc, inplace=True)
df2.drop('NO_CATEGORIA', axis=1, inplace=True)

# Excluindo colunas antigas
df2.drop('NO_MUNICIPIO', axis=1, inplace=True)
df2.drop('NO_REGIAO', axis=1, inplace=True)

# Renomeando as colunas 'NU_ANO_CENSO' em ANO_CENSO e 'CO_MUNICIPIO' em CD_IBGE_MUNICIPIO
df2 = df2.rename(columns={'NU_ANO_CENSO': 'ANO_CENSO'})
df2 = df2.rename(columns={'CO_MUNICIPIO': 'CD_IBGE_MUNICIPIO'})

df2
```

```python
# Listas de prefixos e sufixos
antigos_prefixos = ['FUN_CAT', 'FUN_AI_CAT', 'FUN_AF_CAT', 'MED_CAT']
novos_prefixos = ['EF', 'EF_AI', 'EF_AF', 'EM']
sufixos = ['N1', 'N2', 'N3', 'N4', 'N5', 'N6']
antigos_sufixos = ['1', '2', '3', '4', '5', '6']

# Loop para adicionar as novas colunas com os sufixos
for prefixo in antigos_prefixos:
    for antigo, novo in zip(antigos_sufixos, sufixos):
        # Adiciona a nova coluna com o sufixo "N" usando o novo prefixo
        df2[f'{novos_prefixos[antigos_prefixos.index(prefixo)]}_{novo}'] = df2[f'{prefixo}_{antigo}']
        
        # Remove a coluna antiga
        df2.drop(f'{prefixo}_{antigo}', axis=1, inplace=True)

# Lista de colunas a serem removidas (N3 a N6 e FUN_CAT_3 a FUN_CAT_6, FUN_AI_N3 a FUN_AI_N6, FUN_AF_N3 a FUN_AF_N6)
colunas_a_remover = [f'{novos_prefixos[antigos_prefixos.index(prefixo)]}_{novo}' for prefixo in antigos_prefixos for novo in sufixos[2:]]

# Remove as colunas não desejadas
df2.drop(colunas_a_remover, axis=1, inplace=True)

df2
```

## Criação do df3

```python
df3 = pd.concat([df1, df2])
```

## Ajustes e tratamentos de dados do Dataframe df3

```python
# Reordena as colunas do DataFrame de acordo com a nova ordem
nova_ordem = ['ANO_CENSO', 'CD_IBGE_UF', 'CD_IBGE_MUNICIPIO', 'TP_DEPENDENCIA_ADM', 'CD_LOCALIZACAO_ESCOLA',
              'EF_N1', 'EF_N2', 'EF_AI_N1', 'EF_AI_N2', 'EF_AF_N1', 'EF_AF_N2', 'EM_N1', 'EM_N2']

df3 = df3[nova_ordem]
```

```python
# Lista de prefixos e sufixos
prefixos = ['EF', 'EF_AI', 'EF_AF', 'EM']
sufixos = ['N1', 'N2']

# Loop para converter as colunas
for prefixo in prefixos:
    for sufixo in sufixos:
        coluna_a_converter = f'{prefixo}_{sufixo}'
        # Substitui o valor '--' por 0 antes da conversão
        df3[coluna_a_converter] = df3[coluna_a_converter].replace('--', 0).astype(float)
        
# Convertendo a coluna 'CD_IBGE_MUNICIPIO' para valores numéricos, tratando valores não numéricos como NaN
df3['CD_IBGE_MUNICIPIO'] = pd.to_numeric(df3['CD_IBGE_MUNICIPIO'], errors='coerce')
df3['CD_IBGE_MUNICIPIO'].fillna(0, inplace=True)

# Convertendo para inteiro
df3['CD_IBGE_MUNICIPIO'] = df3['CD_IBGE_MUNICIPIO'].astype(int)
df3
```

## Criação do Dataframe final

```python
# Criando uma cópia do DataFrame df3 para cada etapa de ensino
df_ef = df3.copy()
df_ef_ai = df3.copy()
df_ef_af = df3.copy()
df_em = df3.copy()

# Configurando os valores corretos para as colunas CD_ETAPA_ENSINO e VL_INDICADOR nos DataFrame
df_ef['CD_ETAPA_ENSINO'] = 12
df_ef['VL_INDICADOR'] = df_ef['EF_N1'] + df_ef['EF_N2']

df_ef_ai['CD_ETAPA_ENSINO'] = 4
df_ef_ai['VL_INDICADOR'] = df_ef_ai['EF_AI_N1'] + df_ef_ai['EF_AI_N2']

df_ef_af['CD_ETAPA_ENSINO'] = 8
df_ef_af['VL_INDICADOR'] = df_ef_af['EF_AF_N1'] + df_ef_af['EF_AF_N2']

df_em['CD_ETAPA_ENSINO'] = 16
df_em['VL_INDICADOR'] = df_em['EM_N1'] + df_em['EM_N2']

# Removendo as colunas indesejadas em cada DataFrame
colunas_para_remover = ['EF_N1', 'EF_N2', 'EF_AI_N1', 'EF_AI_N2', 'EF_AF_N1', 'EF_AF_N2', 'EM_N1', 'EM_N2']
df_ef.drop(columns=colunas_para_remover, inplace=True)
df_ef_ai.drop(columns=colunas_para_remover, inplace=True)
df_ef_af.drop(columns=colunas_para_remover, inplace=True)
df_em.drop(columns=colunas_para_remover, inplace=True)

# Concatenando os DataFrames em um único DataFrame final (df_final)
df_final = pd.concat([df_ef, df_ef_ai, df_ef_af, df_em], ignore_index=True)

# Ordenando o Dataframe df_final
df_final.sort_values(by=['CD_IBGE_UF', 'CD_IBGE_MUNICIPIO'], inplace=True)

df_final
```
